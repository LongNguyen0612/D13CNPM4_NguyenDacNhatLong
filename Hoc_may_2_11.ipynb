```python
import pandas as pd
#from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import learning_curve
#from sklearn.externals import joblib
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
df = pd.read_excel('sentimentvn.xlsx')
df
```

```python
stopword=[]
stopword.append('.')
stopword.append(',')
stopword.append(';')
#stopword.append('ứng')

print(stopword)
def clean_tweets(df):
    df['text']=df['text'].str.lower()
    return df
```

```python
X=df['text'].to_list()

print('X=',X)

y=df['label'].to_list()
print('y=',y)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=10)
```

```python
Vectorizer = CountVectorizer(analyzer = u'word',ngram_range=(1, 1),stop_words=set(stopword))
TfIdf=Vectorizer.fit(X_train)
X_train=TfIdf.transform(X_train)
X_test=TfIdf.transform(X_test)
print('Ma tran X_train=',X_train.todense())
print('X_train.shape=',X_train.shape)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1500)

print('X_train.todense=',X_train.todense())
X_trainPCA = pca.fit_transform(X_train.todense())

X_testPCA = pca.transform(X_test.todense())
print('X_train_pca=',X_trainPCA)
```

```python
# SỬ DỤNG SVM
print('SVM & PCA')
model =svm.LinearSVC(C=0.1)
model.fit(X_trainPCA,y_train)
```

```python
# evaluation (Bước đánh giá tính hiệu quả của mô hình) 
y_pred=model.predict(X_testPCA)

print('Các độ đo Precision, Reacll, F1-Score đạt được từ mô hình phân lớp SVM: \n',classification_report(y_test, y_pred, digits=4))
print('Ma trận nhầm lẫn đạt được từ mô hình phân lớp SVM: \n',confusion_matrix(y_test, y_pred))
```

```python
class Sequencer():
    
    def __init__(self,
                 all_words,
                 max_words,
                 seq_len,
                 embedding_matrix
                ):
        
        self.seq_len = seq_len
        self.embed_matrix = embedding_matrix
        """
        temp_vocab = Vocab which has all the unique words
        self.vocab = Our last vocab which has only most used N words.
    
        """
        temp_vocab = list(set(all_words))
        self.vocab = []
        self.word_cnts = {}
        """
        Now we'll create a hash map (dict) which includes words and their occurencies
        """
        for word in temp_vocab:
            # 0 does not have a meaning, you can add the word to the list
            # or something different.
            count = len([0 for w in all_words if w == word])
            self.word_cnts[word] = count
            counts = list(self.word_cnts.values())
            indexes = list(range(len(counts)))
        
        # Now we'll sort counts and while sorting them also will sort indexes.
        # We'll use those indexes to find most used N word.
        cnt = 0
        while cnt + 1 != len(counts):
            cnt = 0
            for i in range(len(counts)-1):
                if counts[i] < counts[i+1]:
                    counts[i+1],counts[i] = counts[i],counts[i+1]
                    indexes[i],indexes[i+1] = indexes[i+1],indexes[i]
                else:
                    cnt += 1
        
        for ind in indexes[:max_words]:
            self.vocab.append(temp_vocab[ind])
                    
    def textToVector(self,text):
        # First we need to split the text into its tokens and learn the length
        # If length is shorter than the max len we'll add some spaces (100D vectors which has only zero values)
        # If it's longer than the max len we'll trim from the end.
        tokens = text.split()
        len_v = len(tokens)-1 if len(tokens) < self.seq_len else self.seq_len-1
        vec = []
        for tok in tokens[:len_v]:
            try:
                vec.append(self.embed_matrix[tok])
            except Exception as E:
                pass
        
        last_pieces = self.seq_len - len(vec)
        for i in range(last_pieces):
            vec.append(np.zeros(100,))
        
        return np.asarray(vec).flatten()
```

```python
sequencer = Sequencer(all_words = [token for seq in x_tokenized for token in seq],
              max_words = 1200,
              seq_len = 15,
              embedding_matrix = model.wv
             )

x_vecs = np.asarray([sequencer.textToVector(" ".join(seq)) for seq in x_tokenized])
print(x_vecs.shape)
```

```python
# WORD2VEC 
from sklearn.decomposition import PCA
pca_model = PCA(n_components=50)
pca_model.fit(x_vecs)
print("Sum of variance ratios: ",sum(pca_model.explained_variance_ratio_))
```

```python
x_comps = pca_model.transform(x_vecs)
x_comps.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_comps,y_prep,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Logistic
classifier = LogisticRegression(random_state = 0)

classifier.fit(x_train, y_train)

#predicting results

y_pred = classifier.predict(x_test)
print(classifier.score(x_test, y_test))
print('Các độ đo Precision, Reacll, F1-Score đạt được từ mô hình phân lớp LogisticRegression: \n',classification_report(y_test, y_pred, digits=4))
print('Ma trận nhầm lẫn đạt được từ mô hình phân lớp LogisticRegression: \n',confusion_matrix(y_test, y_pred))
```

```python
# SVM
classifier = svm.LinearSVC(random_state = 0)

classifier.fit(x_train, y_train)

#predicting results

y_pred = classifier.predict(x_test)
print("Score: ", classifier.score(x_test, y_test))
print('Các độ đo Precision, Reacll, F1-Score đạt được từ mô hình phân lớp SVM: \n',classification_report(y_test, y_pred, digits=4))
print('Ma trận nhầm lẫn đạt được từ mô hình phân lớp SVM: \n',confusion_matrix(y_test, y_pred))
```
